```python
#importing the libraries
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuration of plot that we create
```

```python
# importing the data
data = pd.read_csv("C:/Users/ligio/OneDrive/Desktop/Documents/EXCEL WORK/movies.csv")
```

```python
# reading the data
data.head(10)
```

```python
# let's find the missing values
for col in data.columns:
    percent = np.mean(data[col].isnull())
    print('{} - {}%'.format(col,percent))
```

```python
# fixing the missing values
data["budget"].mean()
```

```python
# fixing the missing values

data["gross"].mean()
```

```python
data["budget"] = data["budget"].replace(np.nan,35589876.192650534)
data["gross"] = data["gross"].replace(np.nan,78500541.01778312)
```

```python
#changing the data type of column
data["budget"] = data["budget"].astype('int64')
data["gross"] = data["gross"].astype('int64')
```

```python
# determining the correlation


# Assuming df is your DataFrame
# Drop non-numeric columns or convert them to numeric
numeric_data = data.select_dtypes(include=['number'])

# Calculate correlation
numeric_data.corr(method='pearson') # pearson,kandall,spearman
```

```python
# high correlaation between budget Vs gross
# i am right

correlation_matrix= numeric_data.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True,)
plt.title('Correlation matrix for numeric feature')
plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.show()
```

```python
# converting all object into category

data_numerized = data

for col in data_numerized.columns:
    if(data_numerized[col].dtype == 'object'):
        data_numerized[col]=data_numerized[col].astype('category')
        data_numerized[col]=data_numerized[col].cat.codes

data_numerized.head()
        
        
```

```python
# creating visual to make understand

correlation_matrix= data_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation matrix for numeric feature')
plt.xlabel('Movie feature')
plt.ylabel('Movie feature')
plt.show()
```

```python
# let make more understandable this correlatiob
correlation_mat = data_numerized.corr()
corr_pair = correlation_mat.unstack()
corr_pair.head()
```

```python
sorted_values = corr_pair.sort_values()
sorted_values.head()
```

```python
high_corr = sorted_values[(sorted_values) > 0.5]
high_corr.head()
```
